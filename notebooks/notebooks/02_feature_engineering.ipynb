```python
# Feature Engineering Notebook

import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../data/nifty_merged_5min.csv")

data.head()
```

```python
data["ema_5"] = data["close_spot"].ewm(span=5, adjust=False).mean()
data["ema_15"] = data["close_spot"].ewm(span=15, adjust=False).mean()

data[["ema_5", "ema_15"]].head()
```

```python
data["spot_return"] = data["close_spot"].pct_change()
data["futures_return"] = data["close_fut"].pct_change()
```

```python
data["futures_basis"] = (
    data["close_fut"] - data["close_spot"]
) / data["close_spot"]
```

```python
data["pcr_oi"] = data["put_oi"] / data["call_oi"]
data["pcr_volume"] = data["put_volume"] / data["call_volume"]
```

```python
data["avg_iv"] = (data["call_iv"] + data["put_iv"]) / 2
```

```python
data.dropna(inplace=True)
```

```python
data.to_csv("../data/nifty_features_5min.csv", index=False)
print("Feature engineering completed and saved")
```
